## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

# Question 2

We Load the file as a string, use the json_normalize function to extract 
the column 'mjtheme_namecode', and convert it to its own data frame

```python
import numpy as np

data2 = json.load((open('data/world_bank_projects.json')))
df_mjtheme_namecode = json_normalize(data2, 'mjtheme_namecode')
df_mjtheme_namecode.tail(20)
df_mjtheme_namecode.head(20)

```

We see that each theme has its own unique code ranging from 1 to 11. However, some themes 
corresponding to the various codes are missing from the name 'column. This means that we must 
count by the values in the 'code' column rather than by the values in the 'name' column.

```python
# Let's count the values of the themes
print(df_mjtheme_namecode['name'].value_counts())
```

One can see that there are 122 missing values, meaning that we can not count by 'name' column. 
Instead, we first create the list of codes corresponding to the top 10 major project themes, sorting
them in descending order, and only then print the names of themes corresponding to these codes. In doing so, we replace '' with NaN's first, drop them subsequently, and apply unique method. 

```python
list_of_codes = list(df_mjtheme_namecode['code'].value_counts()[:10].keys())

for code in list_of_codes:
    print(df_mjtheme_namecode.set_index('code').loc[code, 'name'].replace('', np.nan).dropna().unique()[0])
    
```

# Question 3 

To perform the task of this question, we first create the lists of codes and corresponding themes, then make the dictionary in which each code corresponds to the theme it represents. After that we use this dictionary to iterate over the dataframe replacing missing themes.

```python

# Create the list of codes
list_of_codes = list(df_mjtheme_namecode['code'].value_counts().keys())

# Auxilliary data-frame with the rows corresponding to missing themes removed 
df_aux = df_mjtheme_namecode.replace('', np.nan).dropna()

# Create the list of themes
list_of_themes = []
for code in list_of_codes:
    list_of_themes.append(df_aux.set_index('code').loc[code, 'name'].unique()[0])
   
 # Create a dictionary in which each code corresponds to the theme it represents 
dictionary = {k: v for k,v in zip(list_of_codes, list_of_themes)}


# Replace the empty strings with the theme names corresponding to their codes.
for i in range(len(df_mjtheme_namecode.index)):
    if (df_mjtheme_namecode['name'][i] == ''):
        df_mjtheme_namecode['name'][i] = dictionary[df_mjtheme_namecode['code'][i]]


df_mjtheme_namecode.head(50)
```
